```python
# 데이터 전처리
import re
import json
from bs4 import BeautifulSoup
from nltk.corpus import stopwords

from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
train['review']=train['review'].apply(lambda x: BeautifulSoup(x,"html5lib").get_text())
test['review']=test['review'].apply(lambda x: BeautifulSoup(x,"html5lib").get_text())
```

```python
train['review']=train['review'].apply(lambda x: re.sub("[^a-zA-Z]"," ",x))
test['review']=test['review'].apply(lambda x: re.sub("[^a-zA-Z]"," ",x))
```

```python
train.head(3)
```

```python
import nltk

nltk.download('stopwords')

stops = set(stopwords.words("english"))

for i in range(0, 25000):
    review = train.iloc[i, 2]
    review = review.lower().split()
    words = [r for r in review if not r in stops]
    clean_review = ' '.join(words)
    train.iloc[i, 2] = clean_review
```

```python
for i in range(0, 25000):
    review = test.iloc[i, 1]
    review = review.lower().split()
    words = [r for r in review if not r in stops]
    clean_review = ' '.join(words)
    test.iloc[i, 1] = clean_review
```

```python
train['clean_word_n'] = train['review'].apply(lambda x : len(x.split(' ')))
test['clean_word_n'] = test['review'].apply(lambda x : len(x.split(' ')))
```

```python
fig, axe = plt.subplots(1, 1, figsize=(7,5))
sns.boxenplot(x=train['sentiment'], y=train['clean_word_n'], data=train)
```

```python
from keras.preprocessing.text import Tokenizer
tk = Tokenizer()
tk.fit_on_texts(list(train['review'])+list(test['review']))
text_seq_train = tk.texts_to_sequences(train['review'])
text_seq_test = tk.texts_to_sequences(test['review'])
word_index = tk.word_index
```

```python
print('Total word count is : ', len(word_index))
```

```python
data_info={}
data_info['word_index']=word_index
data_info['word_len']=len(word_index)+1
```

```python
fig = plt.figure(figsize=(15,4))

fig.add_subplot(1,2,1)
sns.distplot(pd.Series(text_seq_train).apply(lambda x : len(x)))

fig.add_subplot(1,2,2)
sns.distplot(pd.Series(text_seq_test).apply(lambda x : len(x)))
```

```python
from keras.preprocessing.sequence import pad_sequences
pad_train=pad_sequences(text_seq_train, maxlen=400)
pad_test=pad_sequences(text_seq_test, maxlen=400)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(pad_train, train['sentiment'],
                                                     random_state=77, test_size=0.07,
                                                     stratify=train['sentiment'])
```

```python
len(tk.word_index)
```

```python
# Modeling
from keras import Sequential
from keras.layers import Dense, Embedding, Flatten

model=Sequential()
model.add(Embedding(len(tk.word_index)+1, 65, input_length=400))
model.add(Flatten())
model.add(Dense(2, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',
             metrics=['acc'])
```

```python
model.fit(x_train,y_train, batch_size=128, 
          epochs=10, validation_data=[x_valid,y_valid]) 
```

```python
res=model.predict(pad_test, batch_size=128)
```

```python
res
```

```python
sub['sentiment_pro']=res[:, 1]
sub.loc[sub['sentiment_pro']>=0.5, "sentiment"] = 1
sub.loc[sub['sentiment_pro']<0.5, "sentiment"] = 1
sub=sub[['id', 'sentiment']]
```

```python
sub.to_csv('result.csv', index=False)
```
